## Allowing user to decide which categorical variable to analyze
### Choose a category to analyze

```python
import pandas as pd

# Ask the user for the variable to analyze
variable_to_analyze = input("Enter the name of the categorical variable you want to analyze: ")

# Check if the provided variable exists in the DataFrame
if variable_to_analyze in df.columns:
    category_counts = df[variable_to_analyze].value_counts()  # Count the occurrences of each category
    total_count = len(df)  # Total number of data points

    # Calculate the proportions as percentages
    proportions = (category_counts / total_count) * 100

    # Create a DataFrame to display the proportions
    category_table = pd.DataFrame({
        'Category': category_counts.index,
        'Frequency': category_counts.values,
        'Proportion (%)': proportions
    })

    # Print the table
    print("Table: Proportions for {} (n={})".format(variable_to_analyze, total_count))
    print(category_table)
else:
    print("The provided variable does not exist in the DataFrame.")
```

# Barcharts 

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group the data by year and count the number of school shootings in each year
shootings_per_year = df['year'].value_counts().sort_index()

# Create a bar chart
plt.figure(figsize=(12, 6))
plt.bar(shootings_per_year.index, shootings_per_year.values, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Number of School Shootings')
plt.title('School Shootings per Year')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Group the data by year and state and count the number of school shootings in each year for each state
shootings_per_state_per_year = df.groupby(['year', 'state']).size().unstack().fillna(0)

# Create a bar chart
plt.figure(figsize=(14, 8))
shootings_per_state_per_year.plot(kind='bar', stacked=True, colormap='Set3')
plt.xlabel('Year')
plt.ylabel('Number of School Shootings')
plt.title('School Shootings per State per Year')
plt.legend(title='State', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
```

```python
import matplotlib.pyplot as plt

# Define the race categories you want to analyze
race_categories = ['white', 'black', 'hispanic', 'asian', 'american_indian_alaska_native', 'hawaiian_native_pacific_islander', 'two_or_more']

# Create a scatterplot for each race category
for race_category in race_categories:
    plt.figure(figsize=(8, 6))  # Set the figure size

    # Filter the data for the specific race category
    filtered_data = df[df[race_category] == 1]

    # Create a scatterplot for casualties vs. enrollment for the current race category
    plt.scatter(filtered_data['enrollment'], filtered_data['casualties'], label=race_category)

    # Set axis labels and a title
    plt.xlabel('Enrollment')
    plt.ylabel('Casualties')
    plt.title(f'Scatterplot of {race_category} vs. Casualties')

    # Show a legend
    plt.legend()

    # Display the plot
    plt.show()
```

# Summary

### Our school shootings data set consisted of 50 columns and 387 rows and the earliest recorded shooting in this set was in 1999 with the most recent happening this year in 2023. The columns included things like the school name, date it happened, the state, enrollment of the school, killed, injured, casualties, shooter age, gender, and race, along with many more categories. 
### As we dove into the data, we were able to find some interesting facts and correlations. Just starting with a basic but useful finding, there averaged one person injured and .5 people killed per shooting. As we moved into the correlations, most were fairly low so when analyzing the data we took that into account when looking at “stronger” correlations. This being said, there was a very weak correlation between the age of the shooter and if someone was killed or injured, with correlations of -.07 and .01. There was one of the stronger correlations of .2 when looking at if the shooter was deceased and if they killed anyone. There was also the strongest correlation within the races of the shooter with white people and casualties. Based on our regplot, we also found that there is almost no correlation between teh number of casualties and the enrollment of the school.
### A couple other interesting findings based on some graphs that we had were that the most shootings have happened in recent years. 2022, 2021, 2018, 2023, 2019 are the top five years for the number of shootings, showing an upward trend. Most casualties happened on Tuesdays with the least amount on Thursdays, and around 120 of the shootings were random while around 70 were targeted. And finally we found that most shootings took place in California with 10%.
